```python
#transformation - image 1
import cv2
import numpy as np
from matplotlib import pyplot as plt
from pylab import *
image1 = cv2.imread('data/FIRE1.jpg')
points1 = np.array([[321,279],[760,310],[325,535],[765,504]])
image2 = cv2.imread('data/FIRE1.jpg')
points2 = np.array([[319,281],[760,280],[325,505],[760,504]])
h,status = cv2.findHomography(points1,points2)
finImage = cv2.warpPerspective(image1,h,(image2.shape[1],image2.shape[0]))
fig, axs = subplots(1, 2, figsize=(15, 16))
axs[0].imshow(image1)
axs[0].axis('off')
title('final result')
axs[1].imshow(finImage)
axs[1].axis('off')
plt.show()
```
